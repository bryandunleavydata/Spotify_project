```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
import numpy as np
import pandas as pd
import pickle
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
#display(X.head())
print()
display(X_scaled_df.head())
```

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score



# Fit K-means clustering with k=12
k = 12
kmeans = KMeans(n_clusters=k, random_state=1234)
kmeans.fit(X)

# Save the trained K-means model
filename = "Model/kmeans_" + str(k) + ".pickle"
with open(filename, "wb") as f:
    pickle.dump(kmeans, f)

# Load the trained K-means model
with open(filename, "rb") as f:
    kmeans_model = pickle.load(f)

# Calculate silhouette score for k=12
silhouette = silhouette_score(X, kmeans.predict(X))

print(f"K-means clustering with k={k} completed.")
print("Silhouette Score:", silhouette)
```

```python
# Predict the cluster labels for your dataset
cluster_labels = kmeans_model.predict(X)

# Add the cluster labels as a new column to your dataset
X['clusters'] = cluster_labels

```

```python
import random
import pandas as pd
from sklearn.preprocessing import StandardScaler


def suggest_song():
    # Take user input
    user_input = input("Please enter a song: ")

    # Convert user input to lowercase for case insensitivity
    user_input = user_input.lower()

    # Search for songs using Spotify API
    results = sp.search(q=user_input, limit=5, market="GB")
    tracks = results['tracks']['items']

    # Display the suggested tracks
    for i, track in enumerate(tracks):
        print(f"{i+1}. {track['name']} - {track['artists'][0]['name']}")

    # Ask the user for the intended track
    intended_track = None
    while intended_track is None:
        choice = input("Was any of the above the intended track? (Enter the number or 'no'): ")
        if choice.lower() == "no":
            return suggest_song()
        try:
            track_index = int(choice) - 1
            intended_track = tracks[track_index]
        except (ValueError, IndexError):
            print("Song not found. Please try again.")

    # Get the audio features for the intended track
    track_id = intended_track['id']
    audio_features = sp.audio_features(track_id)[0]

    # Prepare the features for clustering
    feature_names = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                     'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','duration_ms','time_signature']
    features = pd.DataFrame({feature: audio_features[feature] for feature in feature_names}, index=[0])
    features_scaled = scaler.transform(features)

    # Predict the cluster for the user's song
    cluster = kmeans_model.predict(features_scaled)[0]

    # Filter songs from the same cluster as the user's song
    cluster_songs = X[X['clusters'] == cluster]

    # Pick a random song from the cluster and suggest it to the user
    random_song = cluster_songs.sample(n=1)
    random_track_id = random_song['id'].values[0]

    # Generate and display the suggested track URL
    track_url = f"https://open.spotify.com/embed/track/{random_track_id}"
    display(IFrame(src=track_url, width="320", height="80", frameborder="0", allowtransparency="true", allow="encrypted-media"))

# Run the function
suggest_song()
```
